```python
import pandas as pd
import numpy as np
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

**Categorical variables: Location, WindGustDir, WindDir9am, WindDir3pm, RainToday**

```python
data = pd.read_csv('weatherAUS.csv')
data = data[data['RainTomorrow'].isna() == False]
y = data['RainTomorrow']
date = data['Date']
```

```python
location_df = pd.get_dummies(data['Location'])
WindGustDir_df = pd.get_dummies(data['WindGustDir'], dummy_na = True, prefix = 'WindGustDir')
WindDir9am_df = pd.get_dummies(data['WindDir9am'], dummy_na = True, prefix = 'WindDir9am')
WindDir3pm_df = pd.get_dummies(data['WindDir3pm'], dummy_na = True, prefix = 'WindDir3pm')
RainToday_df = pd.get_dummies(data['RainToday'], dummy_na = True, prefix = 'RainToday')
data = data.drop(['RISK_MM', 'RainTomorrow', 'Date', 'Location', 'WindGustDir', 'WindDir9am', 'WindDir3pm', 'RainToday'], axis = 1)
```

```python
data = pd.concat([data, location_df, WindGustDir_df, WindDir9am_df, WindDir3pm_df, RainToday_df], axis = 1)
data
```

```python
imputer = IterativeImputer(random_state=42, max_iter=100)
imputed = imputer.fit_transform(data)
```

```python
df_imputed = pd.DataFrame(imputed, columns=data.columns)
df_imputed
```

```python
df_imputed.to_csv('second_impu.csv')
```
